```python
#Import dependency
import pandas as pd
```

```python
#Import Jan,2020 bike data
jan="202001-citibike-tripdata.csv"
jan_df=pd.read_csv(jan)
jan_df.head()
#As per data, User Type (Customer = 24-hour pass or 3-day pass user; Subscriber = Annual Member)
#As per data, Gender (Zero=unknown; 1=male; 2=female)
```

```python
#Import all rest 11 months data
feb="202002-citibike-tripdata.csv"
feb_df=pd.read_csv(feb)
mar="202003-citibike-tripdata.csv"
mar_df=pd.read_csv(mar)
apr="202004-citibike-tripdata.csv"
apr_df=pd.read_csv(apr)
may="202005-citibike-tripdata.csv"
may_df=pd.read_csv(may)
june="202006-citibike-tripdata.csv"
june_df=pd.read_csv(june)
july="202007-citibike-tripdata.csv"
july_df=pd.read_csv(july)
aug="202008-citibike-tripdata.csv"
aug_df=pd.read_csv(aug)
sep="202009-citibike-tripdata.csv"
sep_df=pd.read_csv(sep)
oct="202010-citibike-tripdata.csv"
oct_df=pd.read_csv(oct) 
nov="202011-citibike-tripdata.csv"
nov_df=pd.read_csv(nov)
dec="202012-citibike-tripdata.csv"
dec_df=pd.read_csv(dec)
```

```python
#Transform the Jan data
#removing missing or null values
#removing duplicates
jan_df.dropna(inplace=True)
jan_df.drop_duplicates(inplace=True)
jan_df
```

```python
#Transform all the months data
#removing missing or null values
#removing duplicates
feb_df.dropna(inplace=True)
feb_df.drop_duplicates(inplace=True)
mar_df.dropna(inplace=True)
mar_df.drop_duplicates(inplace=True)
apr_df.dropna(inplace=True)
apr_df.drop_duplicates(inplace=True)
may_df.dropna(inplace=True)
may_df.drop_duplicates(inplace=True)
june_df.dropna(inplace=True)
june_df.drop_duplicates(inplace=True)
july_df.dropna(inplace=True)
july_df.drop_duplicates(inplace=True)
aug_df.dropna(inplace=True)
aug_df.drop_duplicates(inplace=True)
sep_df.dropna(inplace=True)
sep_df.drop_duplicates(inplace=True)
oct_df.dropna(inplace=True)
oct_df.drop_duplicates(inplace=True)
nov_df.dropna(inplace=True)
nov_df.drop_duplicates(inplace=True)
dec_df.dropna(inplace=True)
dec_df.drop_duplicates(inplace=True)
```

```python
#add a new column 'Month' with a fixed value 'month name' to df
jan_df['Month'] = 'Jan'
feb_df['Month'] = 'Feb'
mar_df['Month'] = 'Mar'
apr_df['Month'] = 'April'
may_df['Month'] = 'May'
june_df['Month'] = 'June'
july_df['Month'] = 'July'
aug_df['Month'] = 'Aug'
sep_df['Month'] = 'Sep'
oct_df['Month'] = 'Oct'
nov_df['Month'] = 'Nov'
dec_df['Month'] = 'Dec'
```

```python
#randomly sampling 500 rows from the original DataFrame for all months
#sampling without replacement- each row in the original DataFrame can only be selected once in the sample.
sample_jan_df=jan_df.sample(n=500, random_state=42, replace=False)
sample_feb_df=feb_df.sample(n=500, random_state=42, replace=False)
sample_mar_df=mar_df.sample(n=500, random_state=42, replace=False)
sample_apr_df=apr_df.sample(n=500, random_state=42, replace=False)
sample_may_df=may_df.sample(n=500, random_state=42, replace=False)
sample_june_df=june_df.sample(n=500, random_state=42, replace=False)
sample_july_df=july_df.sample(n=500, random_state=42, replace=False)
sample_aug_df=aug_df.sample(n=500, random_state=42, replace=False)
sample_sep_df=sep_df.sample(n=500, random_state=42, replace=False)
sample_oct_df=oct_df.sample(n=500, random_state=42, replace=False)
sample_nov_df=nov_df.sample(n=500, random_state=42, replace=False)
sample_dec_df=dec_df.sample(n=500, random_state=42, replace=False)
```

```python
jan_count = sample_jan_df.count()
jan_count
```

```python
july_count = sample_july_df.count()
july_count
```

```python
datas = [sample_jan_df,sample_feb_df,sample_mar_df,sample_apr_df,sample_may_df,sample_june_df,sample_july_df,sample_aug_df,sample_sep_df,sample_oct_df,sample_nov_df,sample_dec_df]

combined_data = pd.concat(datas)
combined_data
```

```python
# convert the dataframe to CSV files
sample_jan_df.to_csv('jan2020.csv', index=False)
sample_feb_df.to_csv('feb2020.csv', index=False)
sample_mar_df.to_csv('mar2020.csv', index=False)
sample_apr_df.to_csv('apr2020.csv', index=False)
sample_may_df.to_csv('may2020.csv', index=False)
sample_june_df.to_csv('june2020.csv', index=False)
sample_july_df.to_csv('july2020.csv', index=False)
sample_aug_df.to_csv('aug2020.csv', index=False)
sample_sep_df.to_csv('sep2020.csv', index=False)
sample_oct_df.to_csv('oct2020.csv', index=False)
sample_nov_df.to_csv('nov2020.csv', index=False)
sample_dec_df.to_csv('dec2020.csv', index=False)
combined_data.to_csv('combined.csv', index=False)
```
